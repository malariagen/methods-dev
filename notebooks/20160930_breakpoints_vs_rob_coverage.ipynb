```python
scratch_dir = '/lustre/scratch109/malaria/rp7/data/methods-dev/analysis/20160926_pf_60_duplications'
results_dir = '/nfs/team112_internal/rp7/data/methods-dev/analysis/20160926_pf_60_duplications'

pf_6_0_bams_fn = "%s/pf_6_0_bams.txt" % results_dir

dup_results_hrp_list_fn = "%s/dup_results_hrp_list.txt" % results_dir
dup_results_6_0_fn = "%s/dup_results_6_0.txt.old3" % results_dir

rob_hmm_results_fn = "%s/all_HMM_mdr1_pm23_final.txt" % results_dir # Email from Rob 21/09/2016 14:05
```

```python
tbl_dup_results_hrp_list = etl.fromtsv(dup_results_hrp_list_fn).convertnumbers()
print(len(tbl_dup_results_hrp_list.data()))
tbl_dup_results_hrp_list
```

```python
tbl_dup_results_6_0 = etl.fromtsv(dup_results_6_0_fn).convertnumbers()
print(len(tbl_dup_results_6_0.data()))
tbl_dup_results_6_0
```

```python
def is_dup(x):
    if np.isnan(x) or x == 0.0:
        return False
    else:
        return True
    
tbl_dup_results_6_0_flags = tbl_dup_results_6_0.convertall(is_dup)
```

```python
tbl_dup_results_6_0_flags
```

```python
for dup in tbl_dup_results_6_0.header()[1:]:
    print(dup, tbl_dup_results_6_0_flags.valuecounts(dup))
```

```python
tbl_combined_results_hrp_list = (
    tbl_dup_results_hrp_list
    .join(tbl_rob_hmm_results, lkey='sample', rkey='Sample')
    .replaceall('NA', 0)
    .selectge('MedCov', 10)
    .selectlt('Ncnv', 6)
)
print(len(tbl_combined_results_hrp_list.data()))
```

```python
tbl_combined_results_6_0 = (
    tbl_dup_results_6_0
    .join(tbl_rob_hmm_results, lkey='sample', rkey='Sample')
    .replaceall('NA', 0)
    .selectge('MedCov', 10)
    .selectlt('Ncnv', 6)
)
print(len(tbl_combined_results_6_0.data()))
tbl_combined_results_6_0
```

```python
breakpoint_not_hmm_hrp_list = (
    tbl_combined_results_hrp_list
    .select(lambda rec: np.all(np.array(rec)[17:].astype(int) <= 1))
    .select(lambda rec: np.any(np.array(rec)[1:15].astype(float) > 0.01))
)
print(len(breakpoint_not_hmm.data()))
```

```python
breakpoint_not_hmm_6_0 = (
    tbl_combined_results_6_0
    .select(lambda rec: np.all(np.array(rec)[21:].astype(int) <= 1))
    .select(lambda rec: np.any(np.array(rec)[1:19].astype(float) > 0.01))
)
print(len(breakpoint_not_hmm_6_0.data()))
```

```python
hmm_not_breakpoint_6_0 = (
    tbl_combined_results_6_0
    .select(lambda rec: np.sum(np.array(rec)[21:].astype(int) > 2) > 2)
    .select(lambda rec: np.all(np.array(rec)[1:19].astype(float) == 0.0))
)
print(len(hmm_not_breakpoint_6_0.data()))
```
