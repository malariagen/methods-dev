```python
manual_inspection_fn = '/home/Documents/projects/HRP/Manual inspection HRP2 20160622.xlsx'

output_dir = '/nfs/team112_internal/rp7/data/Pf/hrp'
manifest_fn = '%s/metadata/hrp_manifest_20160621.txt' % output_dir
duplicates_fn = '%s/metadata/Workbook2.xlsx' % output_dir
genome_fn = "/lustre/scratch109/malaria/pfalciparum/resources/3D7_V3.fasta"
summary_fn = '%s/metadata/hrp_summary_20160623.xlsx' % output_dir
callable_samples_fn = '%s/metadata/callable_samples_20160623.txt' % output_dir
```

```python
import pyfasta
genome = pyfasta.Fasta(genome_fn)
```

```python
genome_length = 0
CHROMOSOMES = ['Pf3D7_%02d_v3' % n for n in range(1, 15)]
for chrom in CHROMOSOMES:
    genome_length = genome_length + len(genome[chrom])

print(genome_length/2)
```

```python
tbl_hrp2 = etl.fromxlsx(manual_inspection_fn, 'HRP2 definite').pushheader(['sample', 'del_type', 'notes'])
tbl_hrp2
```

```python
tbl_hrp3 = etl.fromxlsx(manual_inspection_fn, 'HRP3 definite').pushheader(['sample', 'del_type', 'notes'])
tbl_hrp3
```

```python
tbl_manifest = etl.fromtsv(manifest_fn).convertnumbers()
tbl_manifest
```

```python
tbl_duplicates = etl.fromxlsx(duplicates_fn).cut(['sample', 'prefix'])
print(len(tbl_duplicates.data()))
tbl_duplicates
```

```python
tbl_usable_samples = (
    tbl_manifest
    .selectge('bases_of_10X_coverage', genome_length/2)
    .selectne('bases_of_10X_coverage', 'unknown')
    .selectne('bases_of_10X_coverage', '')
    .selectnotnone('bases_of_10X_coverage')
    .selectne('sub_continent', 'Lab')
    .select(lambda rec: not rec['sample'].startswith('PG')) # PG0120-C and PG0121-C were annotated as KH
    .replace('region_code', 'KH', 'WKH') # 9 samples from Rick Fairhurst which I'm pretty sure of from Pursat
    .selectne('dataset', 'pf3k_pilot_5_0_broad')
    .cut(['sample', 'dataset', 'type', 'region_code', 'country_code', 'country_name', 'sub_continent',
          'sub_continent_name', 'sub_continent_number', 'bases_of_10X_coverage'])
    .leftjoin(tbl_hrp2.cutout('notes'), key='sample')
    .rename('del_type', 'hrp2_del_type')
    .leftjoin(tbl_hrp3.cutout('notes'), key='sample')
    .rename('del_type', 'hrp3_del_type')
    .leftjoin(tbl_duplicates, key='sample')
    .convertnumbers()
    .sort('bases_of_10X_coverage', reverse=True)
    .distinct('prefix')
)
print(len(tbl_usable_samples.data()))
```
