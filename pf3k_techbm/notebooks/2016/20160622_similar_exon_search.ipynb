```python
def find_similar_exons(chr='Pf3D7_08_v3', start=1374236, end=1375084, max_diff=0.2):
    tbl_this_exon = tbl_all_exons.selecteq('Chr', chr).selecteq('Start', start).selecteq('End', end)
    TotalNExon = tbl_this_exon.values('TotalNExon')[0]
    Length = tbl_this_exon.values('Length')[0]
    PropGeneOccupied = tbl_this_exon.values('PropGeneOccupied')[0]
    MeanUniqueness = tbl_this_exon.values('MeanUniqueness')[0]
    GC = tbl_this_exon.values('GC')[0]
    print(TotalNExon, Length, PropGeneOccupied, MeanUniqueness, GC)
    
    tbl_similar = (
        tbl_all_exons
        .selecteq('TotalNExon', TotalNExon)
        .selectge('Length', Length * (1-max_diff))
        .selectle('Length', Length * (1+max_diff))
        .selectge('PropGeneOccupied', PropGeneOccupied * (1-max_diff))
        .selectle('PropGeneOccupied', PropGeneOccupied * (1+max_diff))
        .selectge('MeanUniqueness', MeanUniqueness * (1-max_diff))
        .selectle('MeanUniqueness', MeanUniqueness * (1+max_diff))
        .selectge('GC', GC * (1-max_diff))
        .selectle('GC', GC * (1+max_diff))
    )
    print(len(tbl_similar.data()))
    
    return(tbl_similar)
     
find_similar_exons(max_diff=0.315).displayall()
    
```
