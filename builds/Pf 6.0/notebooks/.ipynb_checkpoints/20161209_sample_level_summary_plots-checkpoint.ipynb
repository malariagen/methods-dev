```python
tbl_sample_metadata = collections.OrderedDict()
df_sample_metadata = collections.OrderedDict()
for release in sample_metadata_fn:
    tbl_sample_metadata[release] = (
        etl.fromtsv(sample_metadata_fn[release])
        .addfield('continent', lambda rec: continents[rec['study']])
        .addfield('continent_study', lambda rec: "%s__%s" % (rec['continent'], rec['study']))
        .convertnumbers()
        .sort(['continent_study', 'sample'])
    )
    df_sample_metadata[release] = tbl_sample_metadata[release].todataframe()
    print(len(tbl_sample_metadata[release].data()))
    tbl_sample_metadata[release].display(index_header=True)

```

```python
good_studies = collections.OrderedDict()
for release in df_sample_metadata:
    good_samples_per_study = pd.value_counts(df_sample_metadata['pf_60']['study'][df_sample_metadata['pf_60']["pc_pass_missing"] <= 0.05])
    good_studies[release] = list(good_samples_per_study[good_samples_per_study >= 15].keys())
```

```python
good_studies = list(good_samples_per_study[good_samples_per_study >= 15].keys())

for release in df_sample_metadata:
    df = df_sample_metadata[release][
        (df_sample_metadata[release]["pc_pass_missing"] <= 0.05) &
        (np.in1d(df_sample_metadata[release]["study"], good_studies))
    ]
#     for variable in list(df_sample_metadata[release].columns.values):
    for variable in ['pc_bi_nonsynonymous']:
        if not variable in ['sample', 'study', 'center_name', 'source_code', 'run_accessions',
                           'continent', 'continent_study']:
            print(release, variable)
            fig = plt.figure(figsize=(11.69, 11.69/2))
            ax = fig.add_subplot(1, 1, 1)
            sns.boxplot(x="continent_study", y=variable, data=df, ax=ax)
            for item in ax.get_xticklabels():
                item.set_rotation(90)
            if release == 'pf_60':
                ax.set_ylim([0.65, 0.72])
            else:
                ax.set_ylim([0.55, 0.62])
            fig.tight_layout()
            fig.savefig("%s/plots_95pc_called_10sample_studies/%s/png/%s__%s.png" % (output_dir, release, release, variable), dpi=150)
            fig.savefig("%s/plots_95pc_called_10sample_studies/%s/pdf/%s__%s.pdf" % (output_dir, release, release, variable))
#             plt.close(fig)
```

```python
good_studies = list(good_samples_per_study[good_samples_per_study >= 15].keys())

for release in differentiated_variables_ylims:
    df = df_sample_metadata[release][
        (df_sample_metadata[release]["pc_pass_missing"] <= 0.05) &
        (np.in1d(df_sample_metadata[release]["study"], good_studies))
    ]
    for variable in differentiated_variables_ylims[release]:
        fig = plt.figure(figsize=(11.69, 11.69/2))
        ax = fig.add_subplot(1, 1, 1)
        sns.boxplot(x="continent_study", y=variable, data=df, ax=ax)
        for item in ax.get_xticklabels():
            item.set_rotation(90)
        ax.set_ylim(differentiated_variables_ylims[release][variable])
        fig.tight_layout()
        fig.savefig("%s/differentiated_variables/png/%s__%s.png" % (output_dir, release, variable), dpi=150)
        fig.savefig("%s/differentiated_variables/pdf/%s__%s.pdf" % (output_dir, release, variable))
```
