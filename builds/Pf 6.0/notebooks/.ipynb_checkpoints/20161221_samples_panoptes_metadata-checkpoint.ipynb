```python
sis = '1044_Oct2012'
tbl_sis['1044_Oct2012'] = (
    etl
    .fromxlsx(sample_information_sheets['1044_Oct2012'][0], sample_information_sheets['1044_Oct2012'][1], row_offset=1)
    .rename('Code', 'OxfordSrcCode')
    .rename('Collection date', 'sis_Date_%s' % '1044_Oct2012')
    .rename('Origine', 'sis_Location_%s' % sis)
    .addfield('sis_Year_%s' % '1044_Oct2012', lambda rec: str(rec['sis_Date_%s' % '1044_Oct2012'])[0:4])
    .select(lambda rec: rec['OxfordSrcCode'] is not None and rec['OxfordSrcCode'].startswith('163'))
    .cut(['OxfordSrcCode', 'sis_Date_%s' % sis, 'sis_Year_%s' % '1044_Oct2012', 'sis_Location_%s' % sis])
    .distinct('OxfordSrcCode')
)
print(len(tbl_sis[sis].data()))
print(len(tbl_sis[sis].distinct('OxfordSrcCode').data()))
tbl_sis[sis].valuecounts('sis_Year_%s' % sis).displayall()
```

```python
def is_duplicate(prv, cur, nxt):
    if prv is None and (nxt['pf60_OxfordDonorCode'] == cur['pf60_OxfordDonorCode']):
        return(True)
    elif prv is None and (nxt['pf60_OxfordDonorCode'] != cur['pf60_OxfordDonorCode']):
        return(False)
    elif nxt is None and (prv['pf60_OxfordDonorCode'] == cur['pf60_OxfordDonorCode']):
        return(True)
    elif nxt is None and (prv['pf60_OxfordDonorCode'] != cur['pf60_OxfordDonorCode']):
        return(False)
    elif(
            (prv['pf60_OxfordDonorCode'] == cur['pf60_OxfordDonorCode']) or
            (nxt['pf60_OxfordDonorCode'] == cur['pf60_OxfordDonorCode'])
        ):
        return(True)
    else:
        return(False)
    
def discard_as_duplicate(prv, cur, nxt):
    if prv is None:
        return(False)
    elif (prv['pf60_OxfordDonorCode'] == cur['pf60_OxfordDonorCode']):
        return(True)
    else:
        return(False)
    

```
