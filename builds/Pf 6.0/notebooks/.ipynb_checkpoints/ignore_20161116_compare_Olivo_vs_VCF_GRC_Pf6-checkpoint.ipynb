```python
def show_rg(sample='PG0282'):
    line = !grep {sample} {bwa_aln_fofn}
    bam_fn = line[0].split('\t')[0]
    rg = !samtools view -H {bam_fn} | grep RG
    print(rg[0])
    line = !grep {sample} {bwa_mem_fofn}
    bam_fn = line[0].split('\t')[0]
    rg = !samtools view -H {bam_fn} | grep RG
    print(rg[0])
```
