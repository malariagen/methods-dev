```python
def which_release(rec):
    if rec['in_5_0']:
        if (not rec['in_5_1']) or (not rec['in_6_0']):
            return('5_0_NOT_SUBSEQUENT')
        else:
            return('5_0')
    elif rec['in_5_1']:
        if (not rec['in_6_0']):
            return('5_1_NOT_SUBSEQUENT')
        else:
            return('5_1')
    elif rec['in_6_0']:
        return('6_0')
    elif rec['State'] == 'in progress':
        return('in_progress')
    else:
        return('no_release')
```

```python
def which_release(rec):
    if rec['pf60_InV5']:
        if (not rec['pf60_InV5.1']) or (not rec['in_6_0']):
            return('5_0_NOT_SUBSEQUENT')
        else:
            return('5_0')
    elif rec['pf60_InV5.1']:
        if (not rec['in_6_0']):
            return('5_1_NOT_SUBSEQUENT')
        else:
            return('5_1')
    elif rec['in_6_0']:
        return('6_0')
    elif rec['State'] == 'in progress':
        return('in_progress')
    else:
        return('no_release')
```
