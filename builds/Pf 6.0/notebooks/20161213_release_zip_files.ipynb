```python
output_dir = '/lustre/scratch109/malaria/rp7/data/methods-dev/builds/Pf6.0/20161213_release_zip_files'
!mkdir -p {output_dir}/scripts
!mkdir -p {output_dir}/log

release_package_dir = collections.OrderedDict()
sample_metadata_fn = collections.OrderedDict()
release_package_dir['pf_60'] = '/nfs/team112_internal/production/release_build/Pf/6_0_release_packages'
release_package_dir['pv_30'] = '/nfs/team112_internal/production/release_build/Pv/3_0_release_packages'
sample_metadata_fn['pf_60'] = '%s/Pf_60_sample_metadata.txt' % release_package_dir['pf_60']
sample_metadata_fn['pv_30'] = '%s/Pv_30_sample_metadata.txt' % release_package_dir['pv_30']

process_report_fn = collections.OrderedDict()
process_report_fn['pf_60'] = 'processReport-Pf_60-20161213.pdf'
process_report_fn['pv_30'] = 'processReport-Pv_30-20161213.pdf'

readme_fn = collections.OrderedDict()
readme_fn['pf_60'] = 'Pf 6.0 README.txt'
readme_fn['pv_30'] = 'Pv 3.0 README.txt'

zip_fn = collections.OrderedDict()
zip_fn['pf_60'] = 'Pf.6.0.20161213.zip'
zip_fn['pv_30'] = 'Pv.3.0.20161213.zip'

study_manifest_fn = collections.OrderedDict()

run_create_zipfile_job_fn = "%s/scripts/run_create_zipfile_job.sh" % output_dir
```

```python
tbl_sample_metadata = collections.OrderedDict()
for release in sample_metadata_fn:
    tbl_sample_metadata[release] = (
        etl.fromtsv(sample_metadata_fn[release])
        .convertnumbers()
    )
    print(len(tbl_sample_metadata[release].data()))
    tbl_sample_metadata[release].display(index_header=True)
```

```python
studies = collections.OrderedDict()
# for release in ['pf_60']:
for release in tbl_sample_metadata:        
    study_manifest_fn = "%s/studies_%s.txt" % (output_dir, release)
    fo = open(study_manifest_fn, 'w')
    studies[release] = tbl_sample_metadata[release].distinct('study').values('study').array()
    for study in studies[release]:
        print(release, study)
        !cp {release_package_dir[release]}/{process_report_fn[release]} \
        {release_package_dir[release]}/{study}/{process_report_fn[release]}
        !cp {release_package_dir[release]}/"{readme_fn[release]}" \
        {release_package_dir[release]}/{study}/"{readme_fn[release]}"
        print(study, file=fo)
    fo.close()
    print()
```

```python
fo = open(run_create_zipfile_job_fn, 'w')
print('''STUDIES_FN=$1
RELEASE_PACKAGE_DIR=$2
ZIP_FN=$3

# LSB_JOBINDEX=1
IN=`sed "$LSB_JOBINDEX q;d" $STUDIES_FN`
read -a LINE <<< "$IN"
STUDY=${LINE[0]}

echo $STUDY

zip $RELEASE_PACKAGE_DIR/$STUDY/$ZIP_FN $RELEASE_PACKAGE_DIR/$STUDY/* -x $RELEASE_PACKAGE_DIR/$STUDY/*.zip -j

''', file=fo)
fo.close()
```

```python
release = 'pf_60'
!LSB_JOBINDEX=16 && bash {run_create_zipfile_job_fn} {"%s/studies_%s.txt" % (output_dir, release)} \
{release_package_dir[release]} {zip_fn[release]}
```
